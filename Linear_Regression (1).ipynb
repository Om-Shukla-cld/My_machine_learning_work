```python
from sklearn.linear_model import LinearRegression

# Training data
# Study hours
X = [[1], [2], [3], [4], [5]]

# Marks
y = [40, 50, 65, 75, 90]

# Model create
model = LinearRegression()

# Train model
model.fit(X, y)

# User input
hours = float(input("Enter how many hours you study: "))

# Prediction
predicted_marks = model.predict([[hours]])

# Output
print(f"Based on your hours {hours}, you may score around {predicted_marks[0]:.2f}")
```

```python
#isbar data userse lunga 
from sklearn.linear_model import LinearRegression

X = []
y = []

n = int(input("Enter number of data points: "))

for i in range(n):
    hours = float(input("Enter study hours: "))
    marks = float(input("Enter marks: "))
    X.append([hours])
    y.append(marks)

model = LinearRegression()
model.fit(X, y)

print("Model trained successfully!")

test_hours = float(input("Enter hours to predict marks: "))
predicted = model.predict([[test_hours]])

print(f"Predicted marks: {predicted[0]:.2f}")
```

```python
#small project

from sklearn.linear_model import LinearRegression

# Training data
X = []  # distance (km)
y = []  # fuel cost (₹)

n = int(input("Enter number of past trips data: "))

for i in range(n):
    dist = float(input("Enter distance (km) for trip " + str(i+1) + ": "))
    cost = float(input("Enter fuel cost (₹) for trip " + str(i+1) + ": "))
    X.append([dist])
    y.append(cost)

# Train model
model = LinearRegression()
model.fit(X, y)

print("\nModel trained successfully!")

# Prediction
future_distance = float(input("Enter distance for future trip (km): "))
predicted_cost = model.predict([[future_distance]])

print("Estimated fuel cost for " + str(future_distance) +
      " km is ₹" + str(round(predicted_cost[0], 2)))
```
