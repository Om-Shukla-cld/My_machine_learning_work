```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

df = pd.read_csv("salary_dataset.csv")
df
```

```python
plt.figure(figsize=(6,4))
plt.hist(df["salary"], bins=10)
plt.title("Salary Distribution (Before Z-score)")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

```python
#calculating z score

df["z_score_salary"] = stats.zscore(df["salary"])
df
```

```python
z_outliers = df[df["z_score_salary"].abs() > 3]
z_outliers
```

```python
df_z_clean = df[df["z_score_salary"].abs() <= 3]
df_z_clean
```
