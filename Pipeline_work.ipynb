```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
import pickle

data = pd.DataFrame({
    "CGPA":[8.0,6.5,7.2,5.5],
    "Skills":["Yes","No","Yes","No"],
    "Package":[1,0,1,0]
})

X = data[["CGPA","Skills"]]
y = data["Package"]

pre = ColumnTransformer([
    ("num", StandardScaler(), ["CGPA"]),
    ("cat", OneHotEncoder(), ["Skills"])
])

pipe = Pipeline([
    ("prep", pre),
    ("model", LogisticRegression())
])

pipe.fit(X,y)
pickle.dump(pipe, open("model.pkl","wb"))   # âœ… full ML system saved
```

```python
model = pickle.load(open("model.pkl","rb"))

new = pd.DataFrame({"CGPA":[7.0],"Skills":["Yes"]})
model.predict(new)   # works perfectly
```
