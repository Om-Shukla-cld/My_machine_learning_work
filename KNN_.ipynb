```python
from sklearn.neighbors import KNeighborsClassifier

# Training data
# [weight in grams, sweetness level]
X = [
    [900, 6],
    [1100, 6.5],
    [1300, 7],
    [1500, 8],
    [1700, 8.5],
    [1900, 9]
]

# Labels
# 0 = Watermelon, 1 = Papaya
y = [0, 0, 0, 1, 1, 1]

# Create model
model = KNeighborsClassifier(n_neighbors=3)

# Train model
model.fit(X, y)

# User input
weight = float(input("Enter fruit weight in grams: "))
sweetness = float(input("Enter sweetness level (1-10): "))

# Prediction
prediction = model.predict([[weight, sweetness]])[0]

if prediction == 0:
    print("Predicted Fruit: Watermelon")
else:
    print("Predicted Fruit: Papaya")
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Training data
# [area in sq ft, number of rooms]
X = [
    [600, 2],
    [750, 2],
    [900, 3],
    [1200, 4],
    [1500, 5],
    [1800, 6]
]

# Labels
# 0 = Flat, 1 = Villa
y = [0, 0, 0, 1, 1, 1]

# Model
model = KNeighborsClassifier(n_neighbors=3)

# Train
model.fit(X, y)

# User input
area = float(input("Enter house area (sq ft): "))
rooms = float(input("Enter number of rooms: "))

# Prediction
prediction = model.predict([[area, rooms]])[0]

if prediction == 0:
    print("Predicted House Type: Flat")
else:
    print("Predicted House Type: Villa")
```
